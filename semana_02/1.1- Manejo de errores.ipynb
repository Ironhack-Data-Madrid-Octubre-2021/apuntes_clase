## Errores de Sintaxis
Este tipo de errores, conocidos tambi칠n como errores sint치cticos o del int칠rprete de Python, son comunes cuando nos iniciamos con un lenguaje de programaci칩n, y m치s a칰n si hemos programado previamente en otros lenguajes. Se deben al desconocimiento u olvido de las reglas l칠xicas que establece Python, y que el int칠rprete detectar치 antes de lanzar la ejecuci칩n del programa. Se trata, por lo tanto, de expresiones mal formadas. Los lenguajes compilados como C realizan un an치lisis del c칩digo para producir un archivo ejecutable, lo que permite detectareste tipo de errores antes de ejecutar el programa. En el caso de los lenguajes interpretados, como es el caso de Python, al no existir dicha fase previa, no tenemos la oportunidad de detectar esos errores. El int칠rprete acepta nuestro c칩digo con la incertidumbre de su correcci칩n, aunque ser치 consciente de ello nada m치s leer el programa, es decir, al lanzarlo y antes de ejecutar cada una de las instrucciones.      
El int칠rprete de Python procesa l칤nea a l칤nea cada instrucci칩n, y si detecta alg칰n error de sintaxis parar치 su proceso, mostrando una peque침a flecha en el lugar donde el error haya sido detectado. Por ejemplo, en el ejemplo anterior:

````python
while True print ("Probando errores")
````

Es de gran ayuda que nos indique el lugar donde ha detectado el error y que, adem치s, nos facilite informaci칩n sobre el mismo. En este caso est치 indicando que la sintaxsis desde el print no es correcta, para que nos demos cuenta de que estamos cometiendo un error al no poner los dos puntos detr치s del True.
En un mensaje de error de sintaxis distinguimos dos partes: 
- La ruta del error: En el ejemplo anterior, todas las l칤neas menos la 칰ltima.
- La descripci칩n del error: en el ejemplo anterior, la 칰ltima l칤nea.         

Los errores de sintaxis normalmente son f치ciles de corregir, pero en ocasiones un error que se desencadena en un punto puede venir de otro punto y esa ruta de error (o traceback) va en orden descendente desde el punto en el cual el error se desencaden칩, pasando por varios puntos intermedios (Si los hay) hasta el punto donde el error ocurri칩. En la pr치ctica, esto quiere decir que si en la l칤nea donde se indica el error no encontramos nada, tendremos que revisar las instrucciones anteriores hasta encontrarlo.

## Excepciones
Una excepci칩n es un error l칩gico que se produce en tiempo de ejecuci칩n. En este caso, la sintaxsis de las instrucciones es correcta y, por este motivo, el int칠rprete de Python no parar치 ni mostrar치 error alguno al lanzar el programa, pero en el momento de ejecutarse se detendr치 y mostrar치 un error, esto es lo que se denomina "error en tiempo de ejecuci칩n".    
Las excepciones van asociadas a distintos tipos, y ese mismo tipo es el que se muestra en el mensaje de error. Adem치s de esta informaci칩n, tambi칠n se muestran detalles que indican la causa del error. Todo esto te permitir치 disponer de informaci칩n suficiente para encontrar el fallo y solucionarlo o gestionarlo.

- ZeroDivisionError: division by zero

##Gesti칩n de excepciones
La gesti칩n o manejo de excepciones es una t칠cnica de programaci칩n para controlar los errores producidos durante la ejecuci칩n de una aplicaci칩n. Se controlan de una forma parecida a una sentencia condicional. Si no se produce una excepci칩n (General o espec칤fica), que ser칤a el caso normal, la aplicaci칩n continua con las siguientes instrucciones y, si se produce una, se ejecutar치n las instrucciones indicadas por el desarrollador para su tratamiento, lo que pueden continuar la aplicaci칩n o detenerla, dependiendo de casso. La gesti칩n de excepciones en Python comienza con una estructura de palabras reservadas, de la siguiente forma:
```python
try:
    instrucciones
except:
    instrucciones si ocurre esa excepci칩n
```

En esta gesti칩n de excepciones el tipo de la excepci칩n puede ser gen칠rico o espec칤fico. Lo mejor para controlar y gestionar lo ocurrido es hacerlo de la manera m치s espec칤fica posible, escribiendo varias cl치usulas except para el mismo try, de modo que cada una gestione un tipo distinto de excepci칩n. Esto es recomendable porque as칤 haremos una gesti칩n m치s precisa de los casos de error detectados.
```python
try: 
    instruciones
    except tipo de la excepci칩n 1:
        instrucciones si ocurre esa excepci칩n 1
    except tipo de la excepci칩n 2:
        instrucciones si ocurre esa excepci칩n
```

## Validaci칩n de datos con `assert`

Python facilita una forma de establecer condiciones de obligado cunmplimiento, es decir, condiciones que debe cumplir un objeto o, de lo contrario, se producir치 una excepci칩n. Es como una especie de "red de seguridad" ante posibles fallos del programador. La instrucci칩n `assert` a침ade controles para la depuraci칩n de un programa. Nos permite expresar una condici칩n que ha de ser cierta siempre, y que, de no serlo interrumpir치 el programa, generando una excepci칩n que controlar llamada AssertionError. La forma de llamar a esta expresi칩n es la siguiente:
```python
assert condici칩n booleana
```
En caso de que la expresi칩n booleana sea verdadera, assert no hace nada. En caso de que sea falsa, dispara una excepci칩n. Veamos un ejemplo para entenderlo.

````python
assert 1 == 1, "la condici칩n es falsa"
````

````python
assert 1 == 2, "La condici칩n es falsa"
````

````python
lista = [1,2,3,4,5]
nueva = []
try:
    for n in lista:
        assert len(lista) > 3#Condici칩n que se tiene que cumplir, cuando sean 3 elementos se pondr치 FALSE
        print(lista) #VAMOS IMPRIMIENDO LA LISTA EN CADA ITERACI칍N 
        nueva.append(lista.pop())
except AssertionError:
    print(3*4)
    print("La condici칩n es falsa")
print(nueva) # PRINT FUERA DEL BUCLE, EST츼 IMPRIMIENDO DESPU칄S DE EJECUTAR TODO EL C칍DIGO ANTERIOR
````

````python
lista = [1,2,3,4,5]
nueva = []
#ASSERT TUNEANDO EL MENSAJE DE ERROR
try:
    for n in lista:
        assert len(lista) > 3, "la condici칩n es falsa" #Condici칩n que se tiene que cumplir, cuando sean 3 elementos se pondr치 FALSE
        print(lista) #VAMOS IMPRIMIENDO LA LISTA EN CADA ITERACI칍N 
        nueva.append(lista.pop())
except AssertionError as err:
    print(type(err))
    print(err)
print(nueva) # PRINT FUERA DEL BUCLE, EST츼 IMPRIMIENDO DESPU칄S DE EJECUTAR TODO EL C칍DIGO ANTERIOR
````

Solicitamos al usuario que introduzca el car치cter c para continuar o el car치cter f para finalizar. Sin embargo, no podemos asegurar que siempre vaya a introducir alguno de estos caracteres. Por ello, creamos nuestra propia excepci칩n, a la cual llamamos MiExcepcion. De esta manera, si el usuario introduce un car치cter distinto a c y f se lanzar치 la excepci칩n creada y se mostrar치 un mensaje de error indicando que el car치cter introducido no es v치lido.

````python
class MiExcepcion(Exception):
    def __init__(self, stringdescriptivo):
        self.valor = stringdescriptivo
    def __str__(self):
        return f"Error {self.valor}"
````

````python
try:
    fin = False
    while not fin:
        entrada = input("Introduzca c para comenzar o f para finalizar ")
        if entrada != "f" and entrada !="c":
            raise MiExcepcion (f"{entrada} no es v치lido")
        elif entrada == "f":
            fin = True
        #EL PROGRAMA HAR칈A COSAS
except MiExcepcion as err:
    print(err)
````
