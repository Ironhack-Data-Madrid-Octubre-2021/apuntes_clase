## Ejercicio  游 (while)
Lee los n칰meros enteros del teclado, hasta que el usuario introduzca el 0. Finalmente, muestra la suma de todos los n칰meros positivos introducidos.

```python
# Tu c칩digo aqu칤
num = int(input("Introduce un n칰mero "))
total = 0
while num != 0:
    print(num)
    total += num
    print(total)
    num = int(input("Introduce un n칰mero "))
print(total)
```
