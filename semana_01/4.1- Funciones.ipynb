Tambi茅n es posible usar estas funciones an贸nimas como argumentos en la llamada a otras funcoines. Por ejemplo, en una llamada a sorted() o list.sort() podemos usar una funci贸n lambda como argumento para el par谩metro key:

````python
distancias = [("Madrid", 331), ("C贸rdoba", 117), ("Barcelona",798)]
````

````python
#Ordenamos por la distancia, de menor a mayor by default 
sorted(distancias, key= lambda x: x[1])
````

````python
#Ordenamos por distancia pero de mayor a menor, con reverse = True
sorted(distancias, key= lambda x: x[1], reverse=True)
````

````python
sorted(distancias)
````

````python
#Ordenamos alfab茅ticamente utilizando el primer elemento x[0]
sorted(distancias, key=lambda x: x[0])
````

Por 煤ltimo, si queremos devolver como resultado de una funci贸n a otra funci贸n podemos usar cualquier funci贸n conocida en el 谩mbito de la funci贸n que realiza el retorno, o usar una funci贸n lambda:

````python
def comparaciones(tipo):
    
    def mayor(a,b):
        if a > b:
            return a
        else:
            return b
        
    if tipo == "Mayor":
        return mayor
    
    elif tipo == "Menor":
        return lambda x,y: x if x < y else y
````

````python
def comparaciones_igual(tipo):    
    if tipo == "Mayor":
        return lambda x,y: x if x > y else y
    
    elif tipo == "Menor":
        return lambda x,y: x if x < y else y
    else: 
        return "Hola"
````

````python
comparaciones("Mayor")(6,5)
````

````python
comparaciones("Menor")(8,3)
````

````python
comparaciones_igual("Mayor")(".","-")
````

##Recursividad
Se denomina llamada recursiva (o recursividad), a aquellas funciones que en su algoritmo, hacen referencia s铆 misma.

Las llamadas recursivas suelen ser muy 煤tiles en casos muy puntuales, pero debido a su gran factibilidad de caer en iteraciones infinitas, deben extremarse las medidas preventivas adecuadas y, solo utilizarse cuando sea estrictamente necesario y no exista una forma alternativa viable, que resuelva el problema evitando la recursividad.

Python admite las llamadas recursivas, permitiendo a una funci贸n, llamarse a s铆 misma, de igual forma que lo hace cuando llama a otra funci贸n.

````python
def jugar(intento = 1):
    respuesta = input("驴De qu茅 color es una naranja?  ")
    if respuesta != "Naranja":
        if intento < 3:
            print("\n Fallaste! Int茅ntalo de nuevo")
            intento +=1
            jugar(intento) # Llamada recursiva
        else:
            print("\n Perdiste")
    else:
        print("M谩quina ")
````

````python
import time
def cuenta_atras(num):
    num -=1
    if num > 0:
        print(num)
        time.sleep(1)
        cuenta_atras(num)
    else:
        print("Booooommmmmmmmmm ヰ")
````
