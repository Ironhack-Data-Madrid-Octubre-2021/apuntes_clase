## Histogramas

Una forma de entender los datos es contar cuntas veces aparece cada valor en el [sample](https://en.wikipedia.org/wiki/Sample_(statistics)#:~:text=In%20statistics%20and%20quantitative%20research,points%2C%20sampling%20units%20or%20observations.).

```python
sample = [1,2,3,4,5,6]
```

```python
# Experimento: tirar un dado 20 veces
exp = random.choices(sample, k=20)
```

```python
# Hacemos una comp dict
count_ = {n:list(exp).count(n) for n in exp}
```

```python
count_
```

Observamos que el mtodo anterior slo es vlido para [variables discretas](http://onlinestatbook.com/glossary/discrete_variables.html). De lo visto hasta ahora, observamos:

* Todo menos el ltimo bin (el ms a la derecha) est medio abierto. En otras palabras, si bins es `[1, 2, 3, 4]`
entonces el primer bin es `[1, 2)` (incluyendo 1, pero excluyendo 2) y el segundo `[2, 3)`.

Esta nueva media no parece muy representativa de la poblacin, verdad? Afortunadamente, tenemos un par de opciones para mejorar las estadsticas en este escenario.

#### Media truncada

[Wikipedia:](https://en.wikipedia.org/wiki/Truncated_mean) Una media truncada o media recortada es una medida estadstica de tendencia central, muy parecida a la media y la mediana. Implica el clculo de la media despus de descartar determinadas partes de una distribucin de probabilidad o muestra en el extremo superior e inferior, y normalmente descartando una cantidad igual de ambos.

**Trimmed mean**

```python
# funcin importada de scipy.stats
print(f'Salario medio recortado de Espaa sin Messi: {np.round(trim_mean(mock_salaries, 0.05), 2)}')
print(f'Salario medio recortado de Espaa sin Messi: {np.round(trim_mean(salarios_con_messi, 0.5), 2)}')
```

#### Mediana

[Wikipedia](https://en.wikipedia.org/wiki/Median): En estadstica y teora de la probabilidad, la mediana es un valor que separa la mitad superior de la mitad inferior de una muestra de datos, una poblacin o una distribucin de probabilidad. Para un conjunto de datos, se puede considerar como el valor "medio". La ventaja bsica de la mediana en la descripcin de los datos en comparacin con la media (a menudo descrita simplemente como el "promedio") es que no est sesgada tanto por una pequea proporcin de valores extremadamente grandes o pequeos, y por lo tanto puede dar una mejor idea de un valor "tpico".

```python
print(f'Mediana del salario de Espaa sin Messi: {np.round(np.median(mock_salaries), 2)}')
print(f'Mediana del salario de Espaa con Messi: {np.round(np.median(salarios_con_messi), 2)}')
```

####  Moda

[Wikipedia:](https://en.wikipedia.org/wiki/Mode_(statistics)) La moda es el valor que aparece con ms frecuencia en un conjunto de valores de datos. Las distribuciones pueden tener ms de una moda (o mximos locales). Estas distribuciones se denominan *multimodales*. 

Para el caso de las variables discretas (o *cuasi* discretas) es una cantidad ms o menos fcil de calcular:

```python
df_titanic.age.mode()
```

Pero, cuando la variable es continua, lo ms probable es que ni siquiera tenga valores repetidos.     
En ese caso debes ajustar tus datos a una distribucin conocida (o a una estimacin) y calcular el mximo de la pdf (ver un [ejemplo](https://stackoverflow.com/questions/59662920/how-to-get-the-mode-of-distribution-in-scipy-stats)).  

```python
pd.Series(mock_salaries).plot.kde();
```

```python
kernel = gaussian_kde(mock_salaries)
height = kernel.pdf(mock_salaries)
mode_value = mock_salaries[np.argmax(height)]
print('Moda sin Messi: ', mode_value)
```

```python
kernel = gaussian_kde(salarios_con_messi)
height = kernel.pdf(salarios_con_messi)
mode_value = mock_salaries[np.argmax(height)]
print('Mode con Messi: ', mode_value)
```

### Varianza y desviacin estndar

Adems de saber dnde est *centrada* una distribucin, tambin es importante saber lo *variable* que son los datos, es decir, lo *dispersos* que estn. En este caso son tiles dos estadsticas, la varianza y la desviacin estndar:

$$Var = \sigma^2 = \frac{\sum (x_i - \mu) ^ 2}{N}$$ 

$$Std = \sigma = \sqrt{Var}$$

De [Wikipedia:](https://en.wikipedia.org/wiki/Variance) En teora de la probabilidad y estadstica, la varianza es la expectativa de la desviacin al cuadrado de una variable aleatoria respecto a su media. Informalmente, mide la distancia de un conjunto de nmeros con respecto a su valor medio. 

(**Nota**: Las expresiones anteriores se refieren a la Varianza de toda una poblacin, donde $\mu$ es la media de la poblacin. Cuando estimamos la Varianza o la std a partir de una muestra, la frmula correcta e *insesgada* es:
$$\hat{Var} = \hat\sigma^2 =\frac{\sum (x_i - \bar{x}) ^ 2}{N-1}$$ 


```python
N = 100
plt.scatter(x=np.random.normal(0,0.5, N),y = np.ones(N), c="black", label="Baja varianza")
plt.scatter(x=np.random.normal(0,2, N),y = np.zeros(N), c="y", label="Alta varianza")
```

Obsrvese que la varianza y las desviaciones tpicas tambin son muy sensibles a los valores atpicos:

```python
print(f'Std de los salarios en Espaa sin Messi: {np.round(np.std(mock_salaries), 2)}')
print(f'Std de los salarios en Espaa con Messi: {np.round(np.std(salarios_con_messi), 2)}')
```

```python
plt.hist(np.random.normal(0, 0.5, N), 
         color='b',
         label='low variance',
         alpha=0.2)

plt.hist(np.random.normal(0, 2, N), 
         color='orange',
         label='high variance',
         alpha=0.2)
plt.legend()
plt.show()
```

## Percentiles y boxplots

Los percentiles (o cuantiles) son otra estadstica til para entender las distribuciones. Segn [Wikipedia](https://en.wikipedia.org/wiki/Percentile):

### Covarianza

[Wikipedia:](https://en.wikipedia.org/wiki/Covariance)

En teora de la probabilidad y estadstica, la covarianza es una medida de la variabilidad conjunta de dos variables aleatorias[1] Si los valores mayores de una variable se corresponden principalmente con los valores mayores de la otra variable, y lo mismo ocurre con los valores menores, (es decir, las variables tienden a mostrar un comportamiento similar), la covarianza es positiva[2] En el caso contrario, cuando los valores mayores de una variable se corresponden principalmente con los valores menores de la otra, (es decir, las variables tienden a mostrar un comportamiento opuesto), la covarianza es negativa.       
**El signo de la covarianza muestra, por tanto, la tendencia en la relacin lineal entre las variables**. 




La frmula de la covarianza entre dos varaibles ($j$, y $k$) es similar a la de la varianza:

$$q_{jk}=\frac{1}{N-1}\sum_{i=1}^{N}\left(  x_{ij}-\bar{x}_j \right)  \left( x_{ik}-\bar{x}_k \right)$$

**Nota:** puedes ver que si $j=k$ la covarianza es slo la varianza?

### Correlacin de Pearson

La magnitud de la covarianza no es fcil de interpretar porque no est normalizada y, por tanto, depende de las magnitudes de las variables. Sin embargo, la versin normalizada de la covarianza, el coeficiente de correlacin de Pearson, muestra por su magnitud la fuerza de la relacin lineal. La frmula es

$$\rho_{X,Y} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$


Donde $\sigma_X$ y $\sigma_Y$ son las respectivas desviaciones estndar.

El coeficiente de correlacin de Pearson es una estadstica que mide la correlacin lineal entre dos variables X e Y. Tiene un valor entre +1 y -1. Un valor de +1 es una correlacin lineal positiva total, 0 es ninguna correlacin lineal y -1 es una correlacin lineal negativa total.


**CUIDADO!** El hecho de que la correlacin (o covarianza) sea cero no significa que no haya patrones

Un coeficiente de correlacin alternativo conocido como [coeficiente de correlacin de Spearman](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) resuelve este problema calculando la correlacin de Pearson para los rangos. El signo de la correlacin de Spearman indica la direccin de la asociacin entre X (la variable independiente) e Y (la variable dependiente). Si Y tiende a aumentar cuando X aumenta, el coeficiente de correlacin de Spearman es positivo. Si Y tiende a disminuir cuando X aumenta, el coeficiente de correlacin de Spearman es negativo. Una correlacin de Spearman de cero indica que no hay tendencia a que Y aumente o disminuya cuando X aumenta. La correlacin de Spearman aumenta en magnitud a medida que X e Y se acercan a ser funciones perfectamente montonas entre s. Cuando X e Y estn perfectamente relacionados de forma montona, el coeficiente de correlacin de Spearman se convierte en 1.

Para ms informacin, consulte [este post](https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8) en towardsdatascience.
